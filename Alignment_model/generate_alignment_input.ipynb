```python
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np

f = open("/media/data2/s_minha/original_data_list.txt")
lines = f.readlines()
#1613397381763-0.jpg 1140 2437 1807 2253 1801 2400 1145 2599 28수 4882

for img in os.scandir('/media/data2/s_minha/yolo_input/val/images/'):
    save_name = ''

    name = img.name
    img_info = []
    for line in lines:
        splitted = line.split()
        if name == splitted[0]:
            img_info = splitted
            break
    print(img_info[0])

    # configuring path information
    img_path = os.path.join('/media/data2/s_minha/yolo_input/val/images/', img.name)

    # reading the image and the coordinates
    img = cv2.imread(img_path)
    h, w, c = img.shape
    
    # 절대좌표
    coord_list = [(float(img_info[1]), float(img_info[2])),
                  (float(img_info[3]), float(img_info[4])),
                  (float(img_info[5]), float(img_info[6])),
                  (float(img_info[7]), float(img_info[8]))]

    # 상대좌표
    lista = [(round(coord_list[0][0]/w, 4), round(coord_list[0][1]/h, 4)),
             (round(coord_list[1][0]/w, 4), round(coord_list[1][1]/h, 4)),
             (round(coord_list[2][0]/w, 4), round(coord_list[2][1]/h, 4)),
             (round(coord_list[3][0]/w, 4), round(coord_list[3][1]/h, 4))]
    
    save_name = str(lista[0][0]) + '_' + str(lista[0][1]) + '__' + str(lista[1][0]) + '_' + str(lista[1][1]) + '__' +str(lista[2][0]) + '_' + str(lista[2][1]) + '__' +str(lista[3][0]) + '_' + str(lista[3][1])+'.jpg'
    
    print(save_name)

    save_path = os.path.join('/media/data2/s_minha/alignment_input_v2/val/', save_name)
    cv2.imwrite(save_path, img)
    
```
